```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def relu(x):
    return np.where(x > 0, x, 0)
```

# ELU

```python
def elu(x, alpha=0.3):
    return np.where(x > 0, x, alpha * (np.exp(x) - 1))
```

```python
x = np.linspace(-10, 10, 400)  # 400 points from -10 to 10
y = elu(x, alpha=0.8)  # Compute y values for each x
y_relu = relu(x)


plt.figure(figsize=(6, 4))  # Optional: specifies the size of the figure
plt.plot(x, y, label='ELU')  # Plot x and y and label the curve
plt.plot(x, y_relu, label='ReLU')
plt.title('Plot of the ELU')  # Add a title
plt.xlabel('x')  # Label the x-axis
plt.ylabel('y')  # Label the y-axis
# plt.grid(True)  # Optionally add a grid
plt.legend()  # Add a legend to explain what the curve represents
plt.show()  # Display the plot
```

## Shifted Sinc Unit

```python
def sinc(z):
    return np.sin(np.pi * z) / (np.pi * z)

def shifted_sinc(x):
    return sinc(x - np.pi)
```

```python
x = np.linspace(-10, 10, 400)  # 400 points from -10 to 10
y = shifted_sinc(x)  # Compute y values for each x


plt.figure(figsize=(6, 4))  # Optional: specifies the size of the figure
plt.plot(x, y, label='Shifted Sinc Unit')  # Plot x and y and label the curve
# plt.title('Plot of Shifted Sinc Unit')  # Add a title
plt.xlabel('x')  # Label the x-axis
plt.ylabel('y')  # Label the y-axis
# plt.grid(True)  # Optionally add a grid
plt.legend()  # Add a legend to explain what the curve represents
plt.show()  # Display the plot
```

# GCU

```python
def gcu(x):
    return x * np.cos(x)
```

```python
x = np.linspace(-10, 10, 400)  # 400 points from -10 to 10
y = gcu(x)  # Compute y values for each x


plt.figure(figsize=(6, 4))  # Optional: specifies the size of the figure
plt.plot(x, y, label='GCU')  # Plot x and y and label the curve
# plt.title('Plot of GCU')  # Add a title
plt.xlabel('x')  # Label the x-axis
plt.ylabel('y')  # Label the y-axis
# plt.grid(True)  # Optionally add a grid
plt.legend()  # Add a legend to explain what the curve represents
plt.show()  # Display the plot
```

# ADA


```python
def ada(x, alpha=0.5):
    return np.maximum(x, 0) * np.exp(-x * alpha)
```

```python
x = np.linspace(-10, 10, 400)  # 400 points from -10 to 10
y = ada(x)  # Compute y values for each x


plt.figure(figsize=(6, 4))  # Optional: specifies the size of the figure
plt.plot(x, y, label='ADA')  # Plot x and y and label the curve
plt.title('Plot of ADA')  # Add a title
plt.xlabel('x')  # Label the x-axis
plt.ylabel('y')  # Label the y-axis
# plt.grid(True)  # Optionally add a grid
plt.legend()  # Add a legend to explain what the curve represents
plt.show()  # Display the plot
```

# Leaky ADA

```python
def leakyada(x, alpha=0.5, leak=0.01):
    return leak * np.min(x, 0) + np.max(x, 0) * np.exp(-x * alpha)
```

```python
x = np.linspace(-10, 10, 400)  # 400 points from -10 to 10
y = leakyada(x, alpha=0.3, leak=0.1)  # Compute y values for each x


plt.figure(figsize=(6, 4))  # Optional: specifies the size of the figure
plt.plot(x, y, label='LeakyADA')  # Plot x and y and label the curve
plt.title('Plot of LeakyADA')  # Add a title
plt.xlabel('x')  # Label the x-axis
plt.ylabel('y')  # Label the y-axis
plt.grid(True)  # Optionally add a grid
plt.legend()  # Add a legend to explain what the curve represents
plt.show()  # Display the plot
```
