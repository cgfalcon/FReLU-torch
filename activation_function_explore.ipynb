```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def relu(x):
    return np.where(x > 0, x, 0)
```

# ELU

```python
def elu(x, alpha=0.3):
    return np.where(x > 0, x, alpha * (np.exp(x) - 1))
```

```python
x = np.linspace(-10, 10, 400)  # 400 points from -10 to 10
y = elu(x, alpha=0.8)  # Compute y values for each x
y_relu = relu(x)


plt.figure(figsize=(6, 4))  # Optional: specifies the size of the figure
plt.plot(x, y, label='ELU')  # Plot x and y and label the curve
plt.plot(x, y_relu, label='ReLU')
plt.title('Plot of the ELU')  # Add a title
plt.xlabel('x')  # Label the x-axis
plt.ylabel('y')  # Label the y-axis
plt.grid(True)  # Optionally add a grid
plt.legend()  # Add a legend to explain what the curve represents
plt.show()  # Display the plot
```
